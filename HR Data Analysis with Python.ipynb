Q.11) Is there any correlation beetween performance Rating and Salary?

```python
corr_df = df[['PerformanceRating', 'ExperienceYears', 'Salary']].corr()
corr_df
```

```python
#showing Coorelation with heatmap 
plt.figure(figsize=(6, 4))

sns.heatmap(
    corr_df,
    annot=True,          # show numbers
    fmt=".2f",           # 2 decimals
    cmap="coolwarm",     # diverging palette
    center=0,            # correct correlation center
    linewidths=0.5
)

plt.title("Correlation Heatmap of Key HR Metrics")
plt.show()
```

Q.15) Identify the Department with the highest attrition rate(Resigned %).

```python
df.head(2)
```

```python
dept_counts= df.groupby('Department')['Status'].agg(total_emp= 'count' , resigned = lambda x: (x == 'Resigned').sum())
dept_counts
```

```python
type(dept_counts)
```

```python
# Calculate attrition rate
dept_counts['attrition_rate_%']= (dept_counts['resigned'] / dept_counts['total_emp']) * 100
```

```python
dept_counts
```

```python
dept_counts.sort_values('attrition_rate_%',ascending = False)
```
